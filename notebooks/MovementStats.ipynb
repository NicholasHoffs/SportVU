```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/merged_csv/0021500001.csv')
df = df[df['EVENT']==17]
df = df[df['PLAYER_ID']==-1]
```

```python
LOC = df[['LOC_X','LOC_Y']]
```

```python
LOC
```

```python
LOC.shape
```

```python
diffx = np.diff(LOC['LOC_X'])
diffy = np.diff(LOC['LOC_Y'])

diffx2 = np.square(diffx)
diffy2 = np.square(diffy)

dist = np.sqrt(diffx2+diffy2)

vel = np.array(dist/.04)
```

```python
def moving_average(a, n=3) :
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return np.array(ret[n - 1:] / n)
```

```python
moving_average(vel).shape
```
